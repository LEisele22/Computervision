```python
import cv2

print("OpenCV Version:", cv2.__version__)
```

```python
# Bild laden
img = cv2.imread("test.jpg")

if img is None:
    print("Bild konnte nicht geladen werden!")
else:
    cv2.imshow("Testbild", img)
    cv2.waitKey(0)      # wartet auf Taste
    cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Konnte Webcam nicht Ã¶ffnen!")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Fehler beim Lesen des Kamerabildes!")
        break

    cv2.imshow("Webcam Test", frame)

    # Mit 'q' verlassen
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
